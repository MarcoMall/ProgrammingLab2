## 1. La legge dei grandi numeri

La legge dei grandi numeri è un teorema fondamentale della teoria della probabilità che indica che se ripetiamo molte volte (tendendo all'infinito) lo stesso esperimento, la frequenza di un certo evento tende ad essere costante.

Vale a dire, **la legge dei grandi numeri indica che se lo stesso test viene eseguito ripetutamente (ad esempio, lancio di una moneta), la frequenza con cui si ripeterà un determinato evento (testa/croce) si avvicinerà a una costante. Questa a sua volta sarà la probabilità che questo evento si verifichi.**

Fonte: https://it.economy-pedia.com/11038376-law-of-the-big-numbers

- Scrivere una funzione che simula il lancio di una moneta per N volte e restituisce la frequenza % di volte che è uscita testa.
- Calcolare la frequenza % con la grandezze del campione (numero di lanci) che varia 100 volte da 10 a 20000. Ottengo quindi 100 valori.
- Fare il plot del Numero medio di teste (y) rispetto alla grandezza del campione (x)





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
def lancio_moneta_n(n):
    v=[random.randint(0,1) for i in range(n)]
    return sum(v)/len(v)

x=np.array([i for i in range(10,20001,200)])
y=np.array([lancio_moneta_n(i) for i in range(10,20001,200)])
df=pd.DataFrame([y,x])
df=df.transpose()
df.columns= ["Percentuale_di_teste","Lanci"]


df.plot.line(x='Lanci',y='Percentuale_di_teste')
plt.title('Monetine wow')
plt.xlabel('Lanci')
plt.ylabel('Percentuale_di_teste')
plt.show()
```

- Quante righe e colonne ha il dataset?

- Riempi i valori mancanti nella colonna 'Embarked' con il valore più frequente
- Rimuovi le righe dove il valore di 'Age' è mancante
- Controlla se ci sono righe duplicate
- Calcola l'età media dei passeggeri per ogni classe (`Pclass`), se ci sono valori mancanti di età riempili con il valore medio
- Visualizza la distribuzione dell'età per classe
- Visualizza la distribuzione dell'età per classe dividendo e mostrando insieme i dati di uomini e donne

```python
# 📦 Caricamento del dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
moda = df['Embarked'].mode()
df['Embarked'] = df['Embarked'].fillna(moda[0])
```

```python
df=df.dropna(subset='Age')
df.info()
```

```python
df.drop_duplicates()
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
sns.barplot(data=df, x='Pclass',y='Age', estimator='mean')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))

barre_m = ax.bar(np.arange(3), df[df['Sex']=='female'].groupby('Pclass')['Age'].mean(), 0.35, label='M', color='blue')
barre_f = ax.bar(np.arange(3)+0.35, df[df['Sex']=='male'].groupby('Pclass')['Age'].mean(), 0.35, label='F', color='pink')
ax.set_xticks([0.2,1.2,2.4],labels=['1','2','3'])
ax.legend()
plt.show()
```

## 3. Analisi del Dataset Iris

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.groupby('species')['species'].count()
```

```python
df.groupby('species')[['petal_length','petal_width']].mean()
```

```python
plt.scatter(df[df['species']=='setosa'].petal_length, df[df['species']=='setosa'].petal_width, color='blue')
plt.scatter(df[df['species']=='versicolor'].petal_length, df[df['species']=='versicolor'].petal_width, color='red')
plt.scatter(df[df['species']=='virginica'].petal_length, df[df['species']=='virginica'].petal_width, color='green')
plt.show()
```

```python
df['petal_area']=df['petal_length']*df['petal_width']
df.head(3)
```

```python
sns.boxplot(data=df, x='species', y='petal_area')
```
